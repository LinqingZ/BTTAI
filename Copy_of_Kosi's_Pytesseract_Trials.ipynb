```python
#for visualizing the predicitions
cv2_img = cv2.imread(g_drive_path_2)
data_result_to_visualize = pytesseract.image_to_data(cv2_img, output_type=Output.DICT)
length = len(data_result_to_visualize['level'])

for i in range(length):
  d = data_result_to_visualize
  (x, y, w, h) = (d['left'][i], d['top'][i], d['width'][i], d['height'][i])
  cv2.rectangle(cv2_img, (x, y), (x + w, y + h), (0, 255, 0), 2)


cv2_imshow(cv2_img)
```
